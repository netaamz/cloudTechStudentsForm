```python
# Import necessary libraries and modules
from ipywidgets.widgets.widget_layout import Layout
import json 
from google.colab import drive
drive.mount('/content/drive')
import IPython
from typing import Text
from ipywidgets import widgets, VBox, HBox

# Load the students JSON data
with open("/content/drive/My Drive/students.json", "r") as fid:
  data=json.load(fid)

# Extract student names and initialize courses list
studentsNames=[]
courses=[]
studentsNames.append(' ')
for name in data['Students']:
  studentsNames.append(name['First name']+' '+name['Last name'])

# Create a dropdown list widget for student names
dropdown_list = widgets.Dropdown(
    value=' ',
    options=studentsNames,
    description='Student :',
    disabled=False
)
# Create a text box widget for displaying student email
email_text = widgets.HTML(
    placeholder='aaa@aaa.com',
    description='Email:',
    disabled=True
)

# Create a courses widget to display courses in an unordered list
courses_widget = widgets.HTML('<ul><li>' + '</li><li>'.join(courses) + '</li></ul>')

# Create an accordion widget for courses
accordion = widgets.Accordion(children=[courses_widget], titles=('HTML', ))
accordion.layout.width = '400px'
accordion.set_title(0, "Courses")

# Create a title widget with custom styling
title_widget = widgets.HTML(value='<h1 style="color: #1A237E; font-family: sans-serif;">Students Info</h1>')

# Create a logo widget with custom styling
logo_widget = widgets.HTML(value='<img src="https://w3.braude.ac.il/wp-content/uploads/2022/12/92349-9_NEW-LOGO-2.png" alt="Logo" style="height: 50px; padding-left: 75px;">')
logo_widget.layout.align_self = 'center'

# Create a title box to display the logo
title_box = HBox([title_widget, logo_widget])
title_box.layout.width = '700px'

# Create a text box widget for updating the TV Show field
tv_show_text = widgets.Text(
    value='',
    placeholder='Enter TV Show',
    description='TV Show:',
    disabled=True
)

# Create a button widget for updating the TV Show field
update_button = widgets.Button(
    description='Update TV Show',
    disabled=True,
    button_style='success', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click me',
    icon='check' # (FontAwesome names without the `fa-` prefix)
)

# Create a button widget for opening the link
link_button = widgets.Button(
    description='Open Link',
    disabled=True,
    button_style='info', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='intresting link to click on',
    icon='check' # (FontAwesome names without the `fa-` prefix)
)

# Create a text box widget for displaying the link
link_widg= widgets.Text(
    value ='',
    placeholder='http',
    description='Link:',
    disabled=True
)

output = widgets.Output()

# Define function to update output when dropdown list value changes
def update_output(change):
    selected_student = change['new']
    link_button.disabled=False
    update_button.disabled=False
    for student in data['Students']:
        if selected_student == student['First name'] + ' ' + student['Last name']:
            with output:
                output.clear_output()
                out.clear_output()
                courses=[]
                #shows email, and allowes to send one
                email_text.value='<a href="mailto:{address}">{address}</a>'.format(address=student['Email'])
                for c in student['Courses']:
                  courses.append(c)
                courses_list = '<ul><li>' + '</li><li>'.join(courses) + '</li></ul>'
                courses_style = 'max-height: 300px; max-width: 300px;'
                courses_html = '<div style="{style}">{list}</div>'.format(style=courses_style, list=courses_list)
                courses_widget = widgets.HTML(courses_html)
                accordion.children = [courses_widget]
                link_widg.value= student['Link']
                tv_show_text.disabled = False
                if 'TV Show' in student:
                    tv_show_text.value= student['TV Show']


out = widgets.Output(layout={'width': '500px'})

# Define function to update TV Show field for selected student
def update_tv_show(b):
    selected_student = dropdown_list.value
    for student in data['Students']:
        if selected_student == student['First name'] + ' ' + student['Last name']:
            student['TV Show'] = tv_show_text.value
            with open("/content/drive/My Drive/students.json", "w") as fid:
              json.dump(data,fid)
            tv_show_text.value = ''
            with out:
                out.clear_output()
                print('TV Show updated successfully for', selected_student)
            break
# Define function to open url link in a new tab
def open_url_in_new_tab(url):
    html = '<script>window.open("{url}", "_blank");</script>'.format(url=url)
    display(widgets.HTML(html))

# Define function to register that the link was clicked
def on_button_click(link_button):
    open_url_in_new_tab(link_widg.value)

# Attach event listeners to widgets
link_button.on_click(on_button_click)
dropdown_list.observe(update_output, names='value')
update_button.on_click(update_tv_show)

# Display widgets and arrange them in boxes
display(title_box, dropdown_list)

# create a vertical box for the email widget and the courses accordion
email_courses_box = VBox([email_text, accordion])

# create a horizontal box for the link widget and the link button
link_box = HBox([link_widg, link_button])

# create a horizontal box for the TV show widget and the update button
tv_show_box = HBox([tv_show_text, update_button])

# create a vertical box for all the widgets
all_widgets_box = VBox([email_courses_box, link_box, tv_show_box])

# display the box of widgets
display(all_widgets_box)

display(out)

```
